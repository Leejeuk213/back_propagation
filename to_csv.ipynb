```python
## csv 제작

from PIL import Image
import numpy as np 
test_path ="./data/test/test"
#classifier = ['t','u','v','w','x','y','z']
pixel_list =[]

for i in range(1,21):
    path = test_path+f'{i}.jpeg'
    print(path)
    im = Image.open(path) 
    pixels = list(im.getdata())
    pixels.append((i-1)%7)
    pixel_list.append(pixels)
    
for i in range(21,141):
    path = test_path+f'{i}.jpg'
    print(path)
    im = Image.open(path) 
    pixels = list(im.getdata())
    pixels.append((i-1)%7)
    pixel_list.append(pixels)

#print(pixel_list)
myArray = np.empty((0,257))  # 0~255는 데이터 256은 라벨 
for pixel in pixel_list:
    myArray = np.append(myArray,np.array([pixel]),axis=0)
myArray = myArray.astype(np.float32)
np.savetxt("test.csv", myArray, delimiter=", ", newline="\n",fmt='%i')
```
